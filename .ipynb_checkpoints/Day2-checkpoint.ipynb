#### Write a program to find the simple interest when the value of principle,rate of interest and time period is given.

```python
def simple_interest(p, t, r):
    return (p*t*r) /100

simple_interest(1000, 5, 50)
```

#### Write a program to find the volume of the cylinder. Also find the cost when ,when the cost of 1litre milk is 40Rs.

```python

import math

def cylinder_volume(radius, height):
    return math.pi * radius**2 * height

def milk_cost(volume, cost_per_litre):
    # Convert volume from cubic centimeters to liters (1 liter = 1000 cubic centimeters)
    volume_in_liters = volume / 1000
    return volume_in_liters * cost_per_litre

# Example usage
radius = float(input("Enter the radius of the cylinder (in cm): "))
height = float(input("Enter the height of the cylinder (in cm): "))
cost_per_litre = 40  # Cost of 1 litre of milk in Rs

volume = cylinder_volume(radius, height)
cost = milk_cost(volume, cost_per_litre)

print(f"The volume of the cylinder is {volume:.2f} cubic centimeters.")
print(f"The cost to fill the cylinder with milk is {cost:.2f} Rs.")
```

### Write  a program that will tell whether the given number is divisible by 3 & 6.

```python
def calculate_angle(hour, minute):
    #ensure that hour is within 0-11
    hour = hour %12
    
    #calculate the positions of the hour and minute hands
    hour_angle = (30*hour) + (0.5*minute)
    minute_angle = 6 * minute
    
    #calculate the diff b/w the two angles
    angle = abs(hour_angle - minute_angle)
    #return the smaller angle of the two possible angles
    return min(angle, 360-angle)

hour = int(input("Enter the hour: "))
minute = int(input("Enter the minute: "))
print(f"The angles at {hour}:{minute} is {calculate_angle(hour, minute)}degrees")
```

#### Given two rectangles, find if the given two rectangles overlap or not. A rectangle is denoted by providing the x and y coordinates of two points: the left top corner and the right bottom corner of the rectangle. Two rectangles sharing a side are considered overlapping. (L1 and R1 are the extreme points of the first rectangle and L2 and R2 are the extreme points of the second rectangle).

` Note: It may be assumed that the rectangles are parallel to the coordinate axis.`

```python
# Check if one rectangle is to the left of the other:
# If the right edge of the first rectangle is to the left of the left edge of the second rectangle, they do not overlap.
# If the right edge of the second rectangle is to the left of the left edge of the first rectangle, they do not overlap.
# Check if one rectangle is above the other:
# If the bottom edge of the first rectangle is above the top edge of the second rectangle, they do not overlap.
# If the bottom edge of the second rectangle is above the top edge of the first rectangle, they do not overlap.
# If neither of these conditions is true, the rectangles overlap.
def overlap(L1, R1, L2, R2):
    #if one rectangle is to the left of the corner
    if L1[0] > R2[0] or L2[0] > R1[0]:
        return False
    
    #if one rectangle is above the other
    if R1[1] > L2[1] or R2[1] > L1[1]:
        return False
    
    return True

L1 = (0, 10)
R1 = (10, 0)
L2 = (5, 5)
R2 = (15, 0)

print(overlap(L1, R1, L2, R2))
```

#### Write a program that will take three digits from the user and add the square of each digit.

```python
def squared_element(num):
    for i in range(len(num)):
        num[i] = num[i] ** 2
    return num    

num = [int(input()) for _ in range(3)]
squared_element(num)
```

#### Write a program that will check whether the number is armstrong number or not.

```python
def is_armstrong(number):
    num_str = str(number)
    num_digits = len(num_str)
    sum_of_powers = sum(int(digit) ** num_digits for digit in num_str)
    return sum_of_powers == number


print(is_armstrong(153))
```

#### Write a program that will take user input of (4 digits number) and check whether the number is narcissist number or not.

```python
def is_narcissitic(number):
    num_str = str(number)
    num_digits = len(num_str)
    sum_of_powers = sum(int(digit) * num_digits for digit in num_str)
    return sum_of_powers == number
number = int(input("Enter 4-digit number "))

if is_narcissitic(number):
    print(f"{number} is a narcissistic Number")
else:
    print("Not a narcissistic number")    
```


#### Write a program that will give you the in hand salary after deduction of HRA(10%),DA(5%),PF(3%), and tax(if salary is between 5-10 lakh–10%),(11-20lakh–20%),(20< _   – 30%)(0-1lakh print k).


```python
def salary(salary):
    hra = salary *0.10
    da = salary *0.5
    pf = salary * 0.3
    taxable_income = salary - (hra + da+pf)
    # Calculate tax based on income slabs
    if taxable_income <= 100000:
        tax = 0
    elif 100000 < taxable_income <= 500000:
        tax = 0
    elif 500000 < taxable_income <= 1000000:
        tax = taxable_income * 0.10
    elif 1000000 < taxable_income <= 2000000:
        tax = taxable_income * 0.20
    else:
        tax = taxable_income * 0.30 
    
    in_hand_salary = salary - (hra + da +pf+tax)
    return in_hand_salary
sa_lary = int(input())
print(f"{salary(sa_lary)}")    
```
