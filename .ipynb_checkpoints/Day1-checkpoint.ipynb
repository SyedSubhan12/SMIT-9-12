#### 1.User will input (3ages).Find the oldest one

```python
user = [int(input("Enter your ages")) for _ in range(3)]

oldest = user[0]

if user[1] > oldest:
    oldest = user[1]
    
if user[2] > oldest:
    oldest = user[2]
print(f"The oldest age is :{oldest}")    
```

#### 2.Write a program that will convert celsius value to fahrenheit

```python
#Write a program that will convert celsius value to fahrenheit
def convert(user):
    return (user * 9/5) + 32
user = int(input("Enter the celsius"))
convert(user)
```

#### 3.User will input (2numbers).Write a program to swap the numbers

```python
#User will input (2numbers).Write a program to swap the numbers
def swap(user):
    x = user[0] = user[1]
    return x
user = [int(input("Enter two numbers: ")) for _ in range(2)]
swap(user)
```

### 4.Write a program that will give you the sum of 3 digits

```python
#Write a program that will give you the sum of 3 digits
def sum(user):
    return user[0] + user[1] + user[2]
user = [int(input()) for _ in range(3)]
sum(user)
```

### 5.Write a program that will reverse a four digit number.Also it checks whether the reverse is true.

```python
# Write a program that will tell whether the number entered by the user is odd or even.
def even_odd(user):
    if user % 2 == 0:
        print("Even")
    else:
        print("odd")
user = int(input())
even_odd(user)            
```

### 7.Write a program that will tell whether the given year is a leap year or not.


```python
# Write a program that will tell whether the given year is a leap year or not.
user = int(input())
if user % 4 == 0:
    print("Leap year")
else:
    print("Nor a leap year") 
```

#### 8.Write a program to find the euclidean distance between two coordinates.


```python
# Write a program to find the euclidean distance between two coordinates.
import math

def euclidean_distance(x1, y1, x2, y2):
    return math.sqrt((x2-x1)**2 + (y2-y1)**2)

x1, y1 = map(float, input())
x2, y2 = map(float, input())
euclidean_distance(x1, y1, x2, y2)
```

#### 9.Write a program that take a user inputr of three angles and will find out whether it can form a triangle or not.


```python
# Function to check if three angles can form a triangle
def can_form_triangle(angle1, angle2, angle3):
    return (angle1 + angle2 + angle3 == 180) and (angle1 > 0) and (angle2 > 0) and (angle3 > 0)

# Taking input for the three angles
angle1 = float(input("Enter the first angle: "))
angle2 = float(input("Enter the second angle: "))
angle3 = float(input("Enter the third angle: "))

# Checking if the angles can form a triangle
if can_form_triangle(angle1, angle2, angle3):
    print("The angles can form a triangle.")
else:
    print("The angles cannot form a triangle.")
```

#### 10.Write a program that will take user input of cost price and selling price and determines whether its a loss or a profit

```python
#Write a program that will take user input of cost price and selling price and determines whether its a loss or a profit
cost_price = int(input("Enter cost"))
sell_price = int(input("Enter the sell"))
if sell_price > cost_price:
    print("Profit")
else:
    print("Loss")    
```
