```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url = "https://realpython.github.io/fake-jobs/"
request = requests.get(url)
request
```

```python
info = BeautifulSoup(request.content, 'html.parser')
info
```

```python
title = info.find(class_='title is-1')
title.text.strip()
```

```python
jobs = info.find_all(class_='column is-half')
jobs
```

```python
jobs_data = []
jobs = info.find_all(class_='column is-half')
for job in jobs:
    job_title = job.find(class_='title is-5')
    subtitle = job.find(class_='subtitle is-6 company')
    location = job.find(class_='location')
    datetime = job.find(class_='is-small has-text-grey')
    data = [job_title.text, subtitle.text, location.text, datetime.text]
    jobs_data.append(data)
print(jobs_data)    
```
