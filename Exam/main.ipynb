## Question no 1:Library Management System using Python OOP 
### Problem Statement: 
Create a simplified Library Management System with Python classes that 
represent a Library, Books, and Users. Your goal is to design the system so that 
it enables basic functionalities, including adding and managing books, 
registering users, and borrowing/returning books


```python
#Library Managment System

#base class
class Book:
    """A simplified Library Management system-->
       Having some Functionalities """
    #Library for books
    library = {"Python":False, 
               "Ml":False,
               "Data Science": False}
    #Initializer
    def __init__(self, title, author, isbn, is_borrow):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.is_borrow = False
    #It tends the user to borrow a book if it is available
    def borrow(self):
        #user --- >input
        user = input("Enter the name of the book:  ")
        #If/else statement to check whether a book is available
        if user in Book.library:
            if Book.library[user] == False:
                #Assigning
                Book.library[user] = True
                self.is_borrow = True
                return "Book has been borrowed successfully"
            else:
                return "This book is already borrowed"
        else:
            return "This book is not in the library"
        
                
    def return_book(self):
        user = input("Enter the name of the book:  ")
        if user in Book.library:
            Book.library[user] = False
            self.is_borrow = False
            return "Thank for your Co-operation"

```

# Digital Acc ---> Inheritting from Book class

```python
class DigitalBook(Book):
    def __init__(self, title, author, isbn, is_borrow, file_format):
        super().__init__(title, author, isbn, is_borrow)
        self.file_format = file_format
    #override borrow method  to print an additional message indicating that the book can be accessed online
    def borrow(self):
        return super().borrow() + " and can be accessed online"
# Audio class

class AudioBook(Book):
    def __init__(self, title, author, isbn, is_borrow, duration):
        super().__init__(title, author, isbn, is_borrow)
        self.duration = duration
    #override borrow method  to print an additional message indicating that the book can be listened to online
    def borrow(self):
        return super().borrow() + " and can be listened to online"


        
```

# User Class 

```python
class User():
    def __init__(self, user_id, name, borrowed_books):
        self.user_id = user_id
        self.name = name
        self.borrowed_books = []
    #borrow_book(book): Adds a book to borrowed_books if it is 
#not already borrowed. Use encapsulation to ensure users 
#cannot modify borrowed_books directly.
    def borrow_book(self, book):
        if book.is_borrow == False:
            self.borrowed_books.append(book)
            book.is_borrow = True
            return "Book has been borrowed successfully"
    def return_book(self, book):
        book = input("Enter the name of the book: ")
        if book in self.borrowed_books:
            self.borrowed_books.remove(book)
            book.is_borrow = False
            return "Book has been returned successfully"

        

```

# Library class

```python
class Library:
    def __init__(self, User, Book, DigitalBook, AudioBook):
        self.users = {}
        self.books = {}
        self.User = User
        self.Book = Book
        self.DigitalBook = DigitalBook
        self.AudioBook = AudioBook
    
    def add_book(self, title, author, isbn, is_borrow=False, file_format=None, duration=None):
        if file_format:
            book = self.DigitalBook(title, author, isbn, is_borrow, file_format)
        elif duration:
            book = self.AudioBook(title, author, isbn, is_borrow, duration)
        else:
            book = self.Book(title, author, isbn, is_borrow)
        
        self.books[title] = book
        return book
    # Under construction
    def return_book(self, book):
        self.User.return_books(book)

              
        
    def register_user(self, user_id, name):
        user = self.User(user_id, name, [])
        self.users[user_id] = user
        return user

    def search_book(self, title):
        return self.books.get(title, "Book not found in the library")
    
    def display_available_books(self):
        return {title: book for title, book in self.books.items() if not book.is_borrow}

    def display_borrowed_books(self):
        return {title: book for title, book in self.books.items() if book.is_borrow}

    def display_user_borrowed_books(self, user_id):
        user = self.users.get(user_id, None)
        if user:
            return {user.name: [book.title for book in user.borrowed_books]}
        else:
            return "User not found"
    
    def main(self):
        while True:
            print("\nLibrary Management System")
            print("1. Add Book")
            print("2. Return Book")
            print("3. Search Book")
            print("4. Display Available Books")
            print("5. Display Borrowed Books")
            print("6. Display User Borrowed Books")
            print("7.Register User")
            print("8. Exit")
            choice = int(input("Enter your choice: "))
            
            if choice == 1:
                title = input("Enter the title: ")
                author = input("Enter the author: ")
                isbn = int(input("Enter the ISBN: "))
                is_borrow = bool(input("Is the book borrowed? (True/False): "))
                file_format = input("Enter the file format (if any): ")
                duration = input("Enter the duration (if any): ")
                self.add_book(title, author, isbn, is_borrow, file_format or None, duration or None)
                
            elif choice == 2:
                book = input("Enter the name of the book: ")
                self.return_book(book)
                
            elif choice == 3:
                title = input("Enter the title: ")
                book = self.search_book(title)
                if isinstance(book, str):
                    print(book)
                else:
                    print(f"Title: {book.title}, Author: {book.author}, Borrowed: {book.is_borrow}")

            elif choice == 4:
                available_books = self.display_available_books()
                if available_books:
                    print("Available Books:")
                    for title, book in available_books.items():
                        print(f"{title}: {book.author}")
                else:
                    print("No available books in the library")
            
            elif choice == 5:
                borrowed_books = self.display_borrowed_books()
                if borrowed_books:
                    print("Borrowed Books:")
                    for title, book in borrowed_books.items():
                        print(f"{title}: {book.author}")
                else:
                    print("No borrowed books in the library")
            
            elif choice == 6:
                user_id = int(input("Enter the user ID: "))
                borrowed_books = self.display_user_borrowed_books(user_id)
                if isinstance(borrowed_books, str):
                    print(borrowed_books)
                else:
                    for name, books in borrowed_books.items():
                        print(f"{name}: {books}")
            elif choice == 7:
                user_id = int(input("Enter the user ID: "))
                name = input("Enter the name: ")
                user = self.register_user(user_id, name)
                print(f"User {user.name} has been registered successfully")
                
            elif choice == 8:
                print("Exiting the program")
                break
            else:
                print("Invalid choice, please try again")
                

                
       
    
```

##### File Analysis and Word Search


```python
# #1.  Write a Python function named analyze_file that takes the name of a 
# file as input and returns a dictionary containing the counts of characters, 
# words, and lines in that file.
def analyze_file(file_name):
    with open(file_name, 'r') as file:
        content = file.read()
        char_count = len(content)
        words_count = len(content.split())
        lines_count = content.count('\n') + 1
        return {'characters': char_count, 'words': words_count, 'lines': lines_count}

# 2.  Write another Python function named search_word that takes the name 
# of a file and a word as input and returns the count of occurrences of that 
# word in the file.

def search_words(file_name, word):
    with open(file_name, 'r') as file:
        content = file.read()
        for line in content.split('\n'):
            if word in line.lower():
                print(f"{word} found in line: {line}")
                return content.count(word)
    return 0

analyze_file("example.txt")
search_words("example.txt", "python")
    
```
