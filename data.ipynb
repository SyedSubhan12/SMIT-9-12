```python
import os
import time
import requests
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service  # Import Service

# URL of the Pixabay page for apple images
pixabay_url = 'https://www.pexels.com/search/oranges/'

# Directory to store apple images
output_dir = 'oranges'

# Create the directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Path to your WebDriver (ensure to specify your path here)
webdriver_path = 'chromedriver.exe'  # Update this with the correct path

# Function to download an image
def download_image(image_url, image_name):
    try:
        img_data = requests.get(image_url).content
        with open(image_name, 'wb') as f:
            f.write(img_data)
        print(f'Downloaded {image_name}')
    except Exception as e:
        print(f'Failed to download {image_name}: {e}')

# Initialize WebDriver with Service object (correct initialization)
service = Service(webdriver_path)
driver = webdriver.Chrome(service=service)

# Open the Pixabay page
driver.get(pixabay_url)

# Scroll down the page to load images (adjust the range if more scrolling is needed)
for _ in range(3):  # Adjust range for deeper scrolling if required
    driver.execute_script("window.scrollBy(0, 1000);")
    time.sleep(2)  # Wait for the page to load more images

# Find all image elements on the page
image_elements = driver.find_elements(By.CSS_SELECTOR, 'img')

# Counter for image download limit
image_count = 0

# Loop through image elements and download the first 50 images
for img in image_elements:
    if image_count >= 100:
        break

    # Get the image URL from the src attribute
    img_url = img.get_attribute('src')

    # Skip placeholders or invalid image sources
    if 'https://' not in img_url or 'placeholder' in img_url:
        continue

    # Generate the image name and download it
    image_name = os.path.join(output_dir, f'oranges_{image_count + 1}.jpg')
    download_image(img_url, image_name)

    image_count += 1

# Close the browser
driver.quit()
```
